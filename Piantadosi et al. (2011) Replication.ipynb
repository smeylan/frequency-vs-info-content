# All Languages #

```r

allScores = do.call('rbind', lapply(allResults, function(x){x[['scores']]}))
allDFs = do.call('rbind', lapply(allResults, function(x){x[['df']]}))
```

```r
dictProps = do.call('rbind',lapply(split(allDFs, allDFs$language), function(df){
    inDictionary = (nrow(subset(df, group == 'in dictionary')) / nrow(df))
    outOfDictionary = (nrow(subset(df, group == 'not in dictionary')) / nrow(df))
    english = (nrow(subset(df, group == 'english')) / nrow(df))
    return(data.frame(language=df$language[1], group = c('In Dictionary', 'Not In Dictionary', 'English'
    ), proportion= c(inDictionary, outOfDictionary, english)))    
}))
dictProps$group = factor(dictProps$group, levels=c('English','Not In Dictionary', 'In Dictionary'))

propGroupsPlot = ggplot(dictProps) + geom_bar(aes(x=language, y=proportion, fill=group
), stat='identity') + ylab('Proportion') + xlab('Language') + theme(axis.text.x = element_text(angle = 45, hjust = 1))

pdf(paste0(plotPath,'propGroupsPlot.pdf'), width=8, height=6)
print(propGroupsPlot)
dev.off()
options(repr.plot.width=6, repr.plot.height=4, jupyter.plot_mimetypes = 'image/png')
propGroupsPlot
```

The relationship between frequency and character length is robust in all 3 groups of types, but the relationship between frequency and length is weakened when all three groups are included. Types not in the language are less frequent but more predictable than types in the language. Types from English are approximately as frequent but less predictable than types in the language. 

Why are items not in language / from English shorter?
* Items from English are likely to be higher frequency items from English, and hence generally shorter
* Proper names, acronyms, and nonlinguistic material undergo any morphologial processes are hence generally shorter


#  Binned by Information Content

```r
# # generalize this to building DFs for all languages
allLanguageBins = do.call('rbind',lapply(list(englishResults, spanishResults, frenchResults, portugueseResults, italianResults, germanResults,
    swedishResults, dutchResults, czechResults, polishResults, romanianResults), function(results){
    df = results[['df']]
    uS = getBinnedLengthForPredictor(df,predictor = 'unigramSurprisal',numBins=50)
    tS = getBinnedLengthForPredictor(df,predictor='surprisal',numBins=50)    
    bothS = rbind(cbind(uS, type='Unigram Surprisal'),cbind(tS, type='Trigram Surprisal'))    
    bothS$language = df$language[1]
    return(bothS)
}))    
```
